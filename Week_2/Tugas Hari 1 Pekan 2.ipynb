<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
Series adalah struktur data satu dimensi yang mampu menyimpan berbagai jenis data, namun tipe data dari objek Series harus seragam. Sedangkan list merupakan tipe data yang digunakan untuk mengoleksikan data dalam satu tempat dan memiliki beberapa tipe di dalam datanya. Dictionary adalah tipe data yang berisi sepasang variabel yaitu key: value.

DataFrame merupakan struktur data dua dimensi, sedangkan Series merupakan struktur data satu dimensi.

```python
# code here
pembagian_kota_Bekasi = pd.Series(['Bekasi Utara','Bekasi Timur','Bekasi Selatan','Bekasi Barat'])
display(pembagian_kota_Bekasi)
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
display(diamonds.describe())
print(pd.DataFrame.mean(diamonds['price']))
print(pd.DataFrame.std(diamonds['depth']))
print(pd.DataFrame.max(diamonds['carat']))
```
